```python
def testListLocksEntriesTypeCheckDomains(entry):
    testListLocksEntriesTypeCheckDomainsDict = {
        'domainId': None,
        'name': None,
        'inLockDown': None,
        'correct': None,
    }
    testListLocksEntriesTypeCheckDomainsDict['domainId'] = True if type(entry['domainId']) is int else False
    testListLocksEntriesTypeCheckDomainsDict['name'] = True if type(entry['name']) is str else False
    testListLocksEntriesTypeCheckDomainsDict['inLockDown'] = True if type(entry['inLockDown']) is bool else False
    testListLocksEntriesTypeCheckDomainsDict['correct'] = True if False not in testListLocksEntriesTypeCheckDomainsDict.values() else False

    return testListLocksEntriesTypeCheckDomainsDict
```

```python
def testListLocksEntriesDomains(entry):
    testListLocksEntriesDomainsDict = {
        'domainId': None,
        'name': None,
        'inLockDown': None,
        'typeCheck': None,
        'correctDomain': None,
    }
    testListLocksEntriesDomainsDict['domainId'] = True if 'domainId' in entry else False
    testListLocksEntriesDomainsDict['name'] = True if 'name' in entry else False
    testListLocksEntriesDomainsDict['inLockDown'] = True if 'inLockDown' in entry else False
    testListLocksEntriesDomainsDict['typeCheck'] = testListLocksEntriesTypeCheckDomains(entry)
    testListLocksEntriesDomainsDict['correctDomain'] = True if False not in testListLocksEntriesDomainsDict.values() and testListLocksEntriesDomainsDict['typeCheck']['correct'] == True else False
    return testListLocksEntriesDomainsDict
```

```python
def testListLocksEntriesTypeCheck(entry):
    testListLocksEntriesTypeCheckDict = {
        'lockId': None,
        'isOwner': None,
        'isDeleted': None,
        'lastStatusChange': None,
        'tags': None,
        'domains': None,
        'stakeholders': None,
        'replacementSerialNo': None,
        'replacementRfId': None,
        'type': None,
        'shareStatus': None,
        'commsInterface': None,
        'accessMode': None,
        'rfId': None,
        'description': None,
        'longitude': None,
        'latitude': None,
        'state': None,
        'commissionState': None,
        'safeState': None,
        'inLockDown': None,
        'passFail': None,
    }

    testListLocksEntriesTypeCheckDict['lockId'] = True if type(entry['lockId']) is str else False
    testListLocksEntriesTypeCheckDict['isOwner'] = True if type(entry['isOwner']) is bool else False
    testListLocksEntriesTypeCheckDict['isDeleted'] = True if type(entry['isDeleted']) is bool else False
    testListLocksEntriesTypeCheckDict['lastStatusChange'] = True if type(entry['lastStatusChange']) is int else False
    testListLocksEntriesTypeCheckDict['tags'] = True if type(entry['tags']) is list else False
    testListLocksEntriesTypeCheckDict['domains'] = True if type(entry['domains']) is list else False
    testListLocksEntriesTypeCheckDict['stakeholders'] = True if type(entry['stakeholders']) is list else False
    testListLocksEntriesTypeCheckDict['type'] = True if type(entry['type']) is int else False
    testListLocksEntriesTypeCheckDict['shareStatus'] = True if type(entry['shareStatus']) is str else False
    testListLocksEntriesTypeCheckDict['commsInterface'] = True if type(entry['commsInterface']) is str else False
    testListLocksEntriesTypeCheckDict['accessMode'] = True if type(entry['accessMode']) is int else False
    testListLocksEntriesTypeCheckDict['rfId'] = True if type(entry['rfId']) is int else False
    testListLocksEntriesTypeCheckDict['longitude'] = True if type(entry['longitude']) is float else False
    testListLocksEntriesTypeCheckDict['latitude'] = True if type(entry['latitude']) is float else False
    testListLocksEntriesTypeCheckDict['state'] = True if type(entry['state']) is str else False
    testListLocksEntriesTypeCheckDict['commissionState'] = True if type(entry['commissionState']) is str else False
    testListLocksEntriesTypeCheckDict['safeState'] = True if type(entry['safeState']) is str else False
    testListLocksEntriesTypeCheckDict['inLockDown'] = True if type(entry['inLockDown']) is bool else False
    if 'replacementSerialNo' in entry:
        testListLocksEntriesTypeCheckDict['replacementSerialNo'] = True if type(entry['replacementSerialNo']) is int else False
    if 'replacementRfId' in entry:
        testListLocksEntriesTypeCheckDict['replacementRfId'] = True if type(entry['replacementRfId']) is int else False
    if 'description' in entry:
        testListLocksEntriesTypeCheckDict['description'] = True if type(entry['description']) is str else False
    testListLocksEntriesTypeCheckDict['passFail'] = True if False not in testListLocksEntriesTypeCheckDict.values() else False
    

    return testListLocksEntriesTypeCheckDict
```

```python
def testListLocksEntries(entry):

    testListLocksEntriesDict = {
        'lockId': None,
        'isOwner': None,
        'isDeleted': None,
        'lastStatusChange': None,
        'tags': None,
        'domains': None,
        'stakeholders': None,
        'type': None,
        'shareStatus': None,
        'commsInterface': None,
        'accessMode': None,
        'rfId': None,
        'description': None,
        'longitude': None,
        'latitude': None,
        'state': None,
        'commissionState': None,
        'safeState': None,
        'inLockDown': None,
        'lockReplacedSer': None,
        'lockReplacedRf': None,
        'typeCheck': None,
        'noMissingFields': None,
        'correct': None,
    }

    testListLocksEntriesDict['lockId'] = True if 'lockId' in entry else False
    testListLocksEntriesDict['isOwner'] = True if 'isOwner' in entry else False
    testListLocksEntriesDict['isDeleted'] = True if 'isDeleted' in entry else False
    testListLocksEntriesDict['lastStatusChange'] = True if 'lastStatusChange' in entry else False
    testListLocksEntriesDict['tags'] = True if 'tags' in entry else False
    testListLocksEntriesDict['stakeholders'] = True if 'stakeholders' in entry else False
    testListLocksEntriesDict['type'] = True if 'type' in entry else False
    testListLocksEntriesDict['shareStatus'] = True if 'shareStatus' in entry else False
    testListLocksEntriesDict['commsInterface'] = True if 'commsInterface' in entry else False
    testListLocksEntriesDict['accessMode'] = True if 'accessMode' in entry else False
    testListLocksEntriesDict['rfId'] = True if 'rfId' in entry else False
    testListLocksEntriesDict['longitude'] = True if 'longitude' in entry else False
    testListLocksEntriesDict['latitude'] = True if 'latitude' in entry else False
    testListLocksEntriesDict['state'] = True if 'state' in entry else False
    testListLocksEntriesDict['commissionState'] = True if 'commissionState' in entry else False
    testListLocksEntriesDict['safeState'] = True if 'safeState' in entry else False
    testListLocksEntriesDict['inLockDown'] = True if 'inLockDown' in entry else False
    testListLocksEntriesDict['typeCheck'] = testListLocksEntriesTypeCheck(entry)


    if len(entry) == 18:
        testListLocksEntriesDict['noMissingFields'] = True
    elif len(entry) == 19:
        if 'replacementSerialNo' in entry:
            testListLocksEntriesDict['lockReplacedSer'] = True
            testListLocksEntriesDict['noMissingFields'] = True
        elif 'replacementRfId' in entry:
            testListLocksEntriesDict['lockReplacedRf'] = True
            testListLocksEntriesDict['noMissingFields'] = True
        elif 'description' in entry:
            testListLocksEntriesDict['description'] = True
            testListLocksEntriesDict['noMissingFields'] = True
        else:
            testListLocksEntriesDict['noMissingFields'] = False

    num = 0
    tempDict = {}
    for e in entry['domains']:
        tempDict['d'+str(num)] = testListLocksEntriesDomains(e)
        num += 1
    testListLocksEntriesDict['domains'] = tempDict
    testListLocksEntriesDict['correct'] = True if False not in testListLocksEntriesDict.values() and testListLocksEntriesDict['typeCheck']['passFail'] == True else False

    return testListLocksEntriesDict
```

```python
def testListLocks(lockJSON, count):
    testListLockDict = {
        "noMissingRecords": None,
        "individualRecords": None,
        "recordCheck": None,
        "entriesWithMissing": None,
        "replacedLocksSer": None,
        "replacedLocksRf": None,
        "hasDescription": None,
        "incorrectRecords": None,
    }
    num = 0

    tempDict = {}
    recordCheck = True
    tempMissing = []
    tempReplacedSer = []
    tempReplacedRf = []
    tempDescription = []
    tempIncorrect = []
    for entry in lockJSON:
        temp = testListLocksEntries(entry)
        tempDict["e" + str(num)] = temp
        if temp['noMissingFields'] == False:
            tempMissing.append(num)
        if temp['lockReplacedSer'] == True:
            tempReplacedSer.append(num)
        if temp['lockReplacedRf'] == True:
            tempReplacedRf.append(num)
        if temp['description'] == True:
            tempDescription.append(num)
        if recordCheck:
            if not temp['correct']:
                recordCheck = False
        if not temp['correct']:
            tempIncorrect.append(num)
        num += 1
    testListLockDict['individualRecords'] = tempDict
    testListLockDict['recordCheck'] = recordCheck
    testListLockDict['noMissingRecords'] = True if len(lockJSON) == count else False
    testListLockDict['entriesWithMissing'] = tempMissing
    testListLockDict['replacedLocksSer'] = tempReplacedSer
    testListLockDict['replacedLocksRf'] = tempReplacedRf
    testListLockDict['hasDescription'] = tempDescription
    testListLockDict['incorrectRecords'] = tempIncorrect
    return testListLockDict
```
