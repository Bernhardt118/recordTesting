```python
# Function to create section of the truth dictionary store global details of the entries in list
def testListLocks(lockJSON, count):
    # Creates global section of the list truth dictionary
    testListLockDict = {
        "noMissingRecords": None,
        "individualRecords": None,
        "recordCheck": None,
        "entriesWithMissing": None,
        "replacedLocksSer": None,
        "replacedLocksRf": None,
        "hasNaNvalueLongLat": None,
        "hasDescription": None,
        "incorrectRecords": None,
        "hasNaN": None,
    }
    # Used to add number id key value
    num = 0
    # Creates temporary dictionary to stores individual record in before moving to truth dictionary
    tempDict = {}
    # Used to check if there are any incorrect records
    recordCheck = True
    # Creates temporary list to store id of all records with missing values
    tempMissing = []
    # Creates temporary list to store id of all records that have a replacement serial fields
    tempReplacedSer = []
    # Creates temporary list to store id of all records that have a replacement Rf fields
    tempReplacedRf = []
    # Creates temporary list to store id of all records that have a description fields
    tempDescription = []
    # Creates temporary list to store id of all records that have a description fields
    tempNan = []
    # Used to check if there are any incorrect records
    tempIncorrect = []
    # Loops through all values in the keyHolderJSON file 
    for entry in lockJSON:
        # Passes entry to the testListlocksEntries method and storess return value
        temp = testListLocksEntries(entry)
        # Assigns returned value in the temporary dictionary with entry id
        tempDict["e" + str(num)] = temp
        # Adds entry to list if the record contains missing
        if temp['noMissingFields'] == False:
            tempMissing.append(num)
        # Adds entry to list if the record has replacement serial fields
        if temp['lockReplacedSer'] == True:
            tempReplacedSer.append(num)
        # Adds entry to list if the record has replacement rf fields
        if temp['lockReplacedRf'] == True:
            tempReplacedRf.append(num)
        # Adds entry to list if the record has description fields
        if temp['description'] == True:
            tempDescription.append(num)
        # Finds first incorrect record then stops checking
        if recordCheck:
            if not temp['correct']:
                recordCheck = False
        if not temp['correct']:
            tempIncorrect.append(num)
        if temp['typeCheck']['hasNaNLL']:
            tempNan.append(num)
        num += 1
    # Assigns temporary dictionary to the truth dictionary
    testListLockDict['individualRecords'] = tempDict
    # Assigns temporary list to the truth dictionary
    testListLockDict['recordCheck'] = recordCheck
    # Assigns True value based on if there aren't missing records else False
    testListLockDict['noMissingRecords'] = True if len(lockJSON) == count else False
    # Assigns temporary list to the truth dictionary
    testListLockDict['entriesWithMissing'] = tempMissing
    # Assigns temporary list to the truth dictionary
    testListLockDict['replacedLocksSer'] = tempReplacedSer
    # Assigns temporary list to the truth dictionary
    testListLockDict['replacedLocksRf'] = tempReplacedRf
    # Assigns temporary list to the truth dictionary
    testListLockDict['hasDescription'] = tempDescription
    # Assigns temporary list to the truth dictionary
    testListLockDict['incorrectRecords'] = tempIncorrect
    # Assigns temporary list to the truth dictionary
    testListLockDict['hasNaN'] = tempNan
    # Returns populated dictionary
    return testListLockDict
```
